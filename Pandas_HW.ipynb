```python
#Import dependencies
import pandas as pd
import csv
import numpy as np
```

```python
#Reading the csv file
file_path = "raw_data/schools_complete.csv"
schools_df = pd.read_csv(file_path)
schools_df.head()
```

```python
#Reading the csv file
file_path = "raw_data/students_complete.csv"
students_df = pd.read_csv(file_path)
students_df.head()
```

```python
renamed_schools_df = schools_df.rename(columns={"name":"school", "size": "Total Students", "budget": "Total Budget"})
renamed_schools_df.head()
```

```python
combined_data = pd.merge(students_df,renamed_schools_df, on = "school")
combined_data.head()
```

```python
#Total # of schools
total_schools = combined_data["school"].nunique()
total_schools 
```

```python
#Total # of students
total_students = combined_data["name"].count().sum()
total_students
```

```python
#Total school budget
total_budget = combined_data["Total Budget"].unique().sum()
total_budget
```

```python
#Average math score
avg_math_score = round(combined_data["math_score"].mean(),2)
avg_math_score
```

```python
#Average reading score
avg_reading_score = round(combined_data["reading_score"].mean(),2)
avg_reading_score
```

```python
#total students passing math
total_students_passing_math = combined_data[combined_data["math_score"] >= 65].count()["math_score"]

#% passing math
passing_math = total_students_passing_math/total_students
percentage_passing_math = round(passing_math*100,0)
percentage_passing_math
```

```python
# total students passing reading
total_students_passing_read = combined_data[combined_data["reading_score"]>=65].count()["reading_score"]
#% passing reading
passing_reading = (total_students_passing_read/total_students)
percentage_passing_reading = passing_reading*100
round(percentage_passing_reading, 0)
```

```python
#Overall pass percentage
overall_pass = round((percentage_passing_math + percentage_passing_reading)/2,2)
overall_pass
```

```python

student_summary = pd.DataFrame ([{"Total Schools":total_schools,"Total Students": total_students, "Total School Budget" : total_budget, "Average Math Score": avg_math_score, "Average Reading Score": avg_reading_score,"Percentage Passing Math": percentage_passing_math, "Percentage Passing Reading": percentage_passing_reading, "Overall Passing Rate": overall_pass}])
student_summary
```

# School Summary


```python
set_students_df = schools_df.set_index("name")
clean_students_df = set_students_df.drop(["School ID"],axis = 1)
clean_students_df
```

```python
student_count_per_school = clean_students_df["size"]
student_count_per_school
```

```python
school_type = clean_students_df["type"]
school_type
```

```python
school_budget = clean_students_df["budget"]
school_budget
```

```python
#finding per student budget
budget_per_student = clean_students_df["budget"]/clean_students_df["size"]
budget_per_student.sort_values()
```

```python
#Average Math Score
grouped_data = combined_data.groupby("school")
grouped_data
```

```python
avg_schoolwise_mathscore = grouped_data["math_score"].mean()
avg_schoolwise_mathscore

```

```python
avg_schoolwise_readingscore = grouped_data["reading_score"].mean()
avg_schoolwise_readingscore
```

```python
school_passing_math = combined_data[(combined_data["math_score"] >= 65)]
school_passing_reading = combined_data[(combined_data["reading_score"] >= 65)]
school_passing_math 
```

```python
percentage_passing_math = school_passing_math.groupby(["school"]).count()["name"]/student_count_per_school*100
percentage_passing_math
```

```python
percentage_passing_reading = school_passing_reading.groupby(["school"]).count()["name"]/student_count_per_school*100
percentage_passing_reading
```

```python
overall_passing_rate_schoolwise = (percentage_passing_math + percentage_passing_reading)/2
overall_passing_rate_schoolwise
```

```python
per_school_summary = pd.DataFrame({"School Type": school_type,
                                   "Total Students": student_count_per_school,
                                   "Total School Budget": school_budget,
                                   "Per Student Budget": budget_per_student,
                                   "Average Math Score": avg_schoolwise_mathscore,
                                   "Average Reading Score": avg_schoolwise_readingscore,
                                   "% Passing Math": percentage_passing_math,
                                   "% Passing Reading": percentage_passing_reading,
                                   "% Overall Passing Rate": overall_passing_rate_schoolwise})
```
