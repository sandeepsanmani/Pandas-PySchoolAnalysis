```python
#Import dependencies
import pandas as pd
import csv
import numpy as np
```

```python
#Reading the csv file
file_path = "raw_data/schools_complete.csv"
schools_df = pd.read_csv(file_path)
schools_df.head()
```

```python
#Reading the csv file
file_path = "raw_data/students_complete.csv"
students_df = pd.read_csv(file_path)
students_df.head()
```

```python
fullRecords = pd.merge(schools_df, students_df, left_on='name', right_on='school')
fullRecords
```

```python
avg_reading_math_score = fullRecords.groupby(["school"])['reading_score','math_score'].mean()
avg_reading_math_score
```

```python
pass_percent = fullRecords.groupby(["school"])['reading_score','math_score'].apply(lambda x: x > 65).apply(lambda x: x/x.sum()*100)
pass_percent
```
