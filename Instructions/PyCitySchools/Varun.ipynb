```python
#Reading the csv file
file_path = "raw_data/students_complete.csv"
students_df = pd.read_csv(file_path)
students_df.head()
```

```python
fullRecords = pd.merge(schools_df, students_df, left_on='name', right_on='school')
fullRecords
```

```python
avg_reading_math_score = fullRecords.groupby(["school"])['reading_score','math_score'].mean()
avg_reading_math_score
```
